## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df['line_subtotal'] = df['unit_price'] * df['qty']
df.head()
```

```python
# Function to calculate shipping weight
def shipping_weight(unit_weight,qty):
    return unit_weight * qty

# Function to calculate shipping cost
def shipping_cost(shipping_weight):
    return 7 if shipping_weight > 50 else 10

# Function to calculate shipping price
def shipping_price(unit_weight, qty):
    weight = shipping_weight(unit_weight, qty)
    cost_per_pound = shipping_cost(weight)
    return weight * cost_per_pound

# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.

df['shipping_price'] = df.apply(lambda row: shipping_price(row['unit_weight'], row['qty']), axis=1)
df.head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df['total_price'] = df.apply(lambda row: round((row['line_subtotal'] + row['shipping_price']) * 1.0925, 2), axis = 1)
df.head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
df['line_cost'] = df.apply(lambda row: (row['unit_cost'] * row['qty']) + row['shipping_price'], axis = 1)
df.head()
```

```python
# Create a column for the profit of each line using line cost and line price
df['line_profit'] = df.apply(lambda row: row['total_price'] - row['line_cost'], axis=1)
df.head()
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
top_5_clients_summary = [] # Empty list to hold summary data

for client_id in top_5_clients:
    filtered_lines_client = df[df['client_id'] == client_id] # Filter the data for the client ID
    client_total_units_purchased = filtered_lines_client['qty'].sum() # Sum up the total units purchased
    client_total_shipping_price = filtered_lines_client['shipping_price'].sum() # Sum up the total shipping price
    client_total_revenue = filtered_lines_client['total_price'].sum() # Sum up the total revenue
    client_total_linecost = filtered_lines_client['line_cost'].sum() # Sum up the line cost
    client_total_profit = filtered_lines_client['line_profit'].sum() # Sum up the total profit
    top_5_clients_summary.append({
        'client_id': client_id,
        'qty': client_total_units_purchased,
        'shipping_price': client_total_shipping_price,
        'line_price': client_total_revenue,
        'line_cost': client_total_linecost,
        'line_profit': client_total_profit
    })

top_5_clients_summary_df = pd.DataFrame(top_5_clients_summary)
top_5_clients_summary_df.head()
```

```python
# Format the data and rename the columns to names suitable for presentation.
top_5_clients_summary_df = top_5_clients_summary_df.rename(columns={
    "client_id": "Client ID", 
    "qty": "Units", 
    "shipping_price": "Shipping Price", 
    "line_price": "Total Revenue", 
    "line_cost": "Total Cost", 
    "line_profit": "Total Profit"
})

# Define the money columns. 
money_columns = ["Shipping Price", "Total Revenue", "Total Cost", "Total Profit"]

# Define a function that converts a dollar amount to millions.
def currency_format_millions(value):
    converted_value = value/1000000
    return converted_value

# Apply the currency_format_millions function to only the money columns. 
for column_name in money_columns:
    top_5_clients_summary_df[column_name] = top_5_clients_summary_df[column_name].apply(currency_format_millions)

# Rename the columns to reflect the change in the money format. 
renamed_df_columns = top_5_clients_summary_df.rename(columns={
    "Shipping Price": "Shipping (millions)",
    "Total Revenue": "Total Revenue (millions)",
    "Total Cost": "Total Cost (millions)",
    "Total Profit": "Total Profit (millions)"
})

renamed_df_columns
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
sorted_renamed_df_columns = renamed_df_columns.sort_values("Total Profit (millions)", ascending=False)
print(sorted_renamed_df_columns)
```
